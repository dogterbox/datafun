```python
def to_datanum(df):
    import pandas as pd
    
    _df = df.values.tolist()
    for i in range(len(_df[0])):
        s = []
        for j, _list in enumerate(_df):
            value = _list[i]
            if value not in s:
                s.append(value)
            _df[j][i] = s.index(value)
            
    return pd.DataFrame(_df, columns=list(df.columns.values))
```

## แปลงข้อมูลให้เป็น float nunber
***********************************************
เข้าใจว่า sklearn มันจะ Train ไม่ได้ถ้ามันไม่อยู่ในรูปของ number หล่ะม้างง เพราะฉะนั้นเราต้องแปลง data to num ก่อน ปล.มันมีหลายวิธีการ Train มากกกกกๆๆๆ ขึ้นอยู่กับงานเราด้วย

## สลับข้อมูล
***************************************************
อันนี้เป็นทฤษฏีของ ML เกี่ยวกับเรื่องของการ Train-Test เราควรจะ shuffle ข้อมูลหล่ะม้างง (ต้องรื้อทฤษฏีดูๆ)

```python
df = df.sample(frac=1).reset_index(drop=True)    # shuffle data
```

## แบ่งข้อมูล
Train 70%
Test 30%

***************************************************
อันนี้เป็นทฤษฏีของ ML เกี่ยวกับเรื่องของการ Train-Test (ต้องรื้อทฤษฏีดูๆ)

```python
train_df = df.iloc[:round(len(df)*.7)]    # Train
test_df = df.iloc[round(len(df)*.7)+1:]   # Test
```

## กำหนดส่วนใน dataset ว่า ส่วนไหนเป็นข้อมูล ส่วนไหนเป้าหมาย
X = data/feature/attribute, 
y = taget/class

***********************************************************
ปล. วิธีการจะกำหนดมันก็มีหลายยยยแบบบ stackoverflow รัวๆ

```python
# Train -- ส่วนที่ใช้สอน
X_train = train_df.iloc[:, train_df.columns != 'ท่านเคยซื้อซอฟแวร์ลิขสิทธิ์หรือไม่ (รวมถึงเกมด้วย) ']    # feature
y_train = train_df.iloc[:, train_df.columns == 'ท่านเคยซื้อซอฟแวร์ลิขสิทธิ์หรือไม่ (รวมถึงเกมด้วย) ']    # target

# Test -- ส่วนที่ใช้ทดสอบ
X_test = test_df.iloc[:, test_df.columns != 'ท่านเคยซื้อซอฟแวร์ลิขสิทธิ์หรือไม่ (รวมถึงเกมด้วย) ']    # feature
y_test = test_df.iloc[:, test_df.columns == 'ท่านเคยซื้อซอฟแวร์ลิขสิทธิ์หรือไม่ (รวมถึงเกมด้วย) ']    # target
```

## เลือกอัลกอที่จะใช้สร้างโมเดล
ตัวอย่างเลือกมา 2 อัลกอ คือ knn กับ logistic

```python
from sklearn import neighbors, linear_model

knn = neighbors.KNeighborsClassifier()
logistic = linear_model.LogisticRegression()
```

## สร้างโมเดล (run/fit)
***
ถ้า dataset มันใหญ่ มันก็จะนานเหี้ยๆ

```python
knn_model = knn.fit(X_train, y_train)
logistic_model = logistic.fit(X_train, y_train)
```

## Score
วัดค่าประสิทธิภาพ/ความแม่นยำ-ถูกต้อง ของโมเดลที่เราสร้าง
***
วิธีวัดมันมีหลายตัว ต้องไปรื้อทฤษฏีดู (Math รัวๆ)

```python
print('KNN score: %f' % knn_model.score(X_test, y_test))
print('LogisticRegression score: %f' % logistic_model.score(X_test, y_test))
```

## Save model
***
จะเอาไปใช้ทำอะไรต่อก็ว่ากันไป ทำวงทำเว็ป บลาๆๆๆ

```python
from sklearn.externals import joblib

filename = 'buy_software_copyright_clf.model'
joblib.dump(knn_model, filename)  # save Knn model
```
