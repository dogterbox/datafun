```python
import matplotlib.pyplot as plt
from sklearn import cluster
%matplotlib inline
```

```python
image_path = 'dog.jpg'
image = plt.imread(image_path)
```

```python
plt.figure(figsize = (15, 8))
plt.imshow(image);
```

## Reshap image

```python
image.shape  # h, w, rgb
```

```python
x, y, z = image.shape
image_2d = image.reshape(x*y, z)
image_2d.shape  # h*w, rgb
```

## K-mean clustering
n_clusters = 20  
max_iter = 100  
**cluster_centers is feature 20 dim (column)**

```python
kmeans_cluster = cluster.KMeans(n_clusters=20, max_iter=100)
kmeans_cluster.fit(image_2d)

cluster_centers = kmeans_cluster.cluster_centers_
cluster_labels = kmeans_cluster.labels_
```

```python
cluster_centers  # feature
```

```python
print(*[_ for _ in cluster_labels[:200]], end=' ')  # print cluster_labels sample
```

# Show image
- Normalize cluster centers value to 0-1  
OR  
- Convert cluster centers value to int (0-255)

## Normalize cluster centers value to 0-1

```python
plt.figure(figsize = (15, 8))
plt.imshow((cluster_centers[cluster_labels] / 255).reshape(x, y, z));
```

## Convert cluster centers value to int (0-255)

```python
plt.figure(figsize = (15, 8))
plt.imshow(cluster_centers[cluster_labels].astype(int).reshape(x, y, z));
```
